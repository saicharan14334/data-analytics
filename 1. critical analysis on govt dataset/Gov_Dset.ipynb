# Introduction about Dataset:
              

   **This dataset gives information about tobacco use in different states and union territories of India. It includes details like how many people use tobacco, how many smoke cigarettes, and how many know about programs that help people quit tobacco** 

**Importing Libraries**

# Overview of Dataset:

#### The dataset has 36 rows (one for each state or union territory).
#### There are 5 columns, each containing different types of data.
#### Data Types:
         object (text) → State/UT, Area
         float64 (decimal numbers) → Tobacco-related percentages**
#### No missing values (Non-Null Count is 36 for all columns).**
#### The dataset takes 1.5 KB of memory.**

# Main Features of the Dataset:

#### State/UT – Name of the state or union territory in India.
##### Current Tobacco Users (%) – Percentage of people in the state who currently use any form of tobacco.
#### Current Cigarette Users (%) – Percentage of people in the state who smoke cigarettes.
##### Tobacco Cessation Awareness (%) – Percentage of people aware of tobacco cessation programs (help to quit smoking).
#### Area – Categorizes states into Rural, Urban, or Mixed areas based on their population distribution.

#### ðŸŽ¯ **Dataset Overview:**
The dataset contains **marks of students (n = 86)** across multiple questions in what appears to be a class test or exam. Each column represents either a **specific question or total marks** scored by each student.

#### ðŸ“Š **Columns Description:**
- `Total`: Total marks scored by a student.
- `Q1aM4`, `Q1bM6`, `Q2aM6`, ..., `Q6bM6`:  
  
For example:
- `Q1aM4` â†’ Question 1, part a, max marks = 4
- `Q3bM5` â†’ Question 3, part b, max marks = 5
- `Q5M10` â†’ Whole Question 5, max marks = 10

This structure implies a **sectioned exam** with multiple parts, where students could attempt different sections, and not all attempted every question (many missing values).

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("question_paper.csv")
df
```

```python
#TOTAL values are sorted in ascending order
DF=df.sort_values("Total",ascending=True)
DF
```

```python
#New Columns Q1,Q2,Q3,Q4,Q6 are created above to do analysis.
DF['Q1']=DF["Q1aM4"]+DF["Q1bM6"]
DF['Q2']=DF['Q2aM6']+DF['Q2bM4']
DF['Q3']=DF['Q3aM5']+DF['Q3bM5']
DF['Q4']=DF['Q4aM3']+DF['Q4bM7']
DF['Q5']=DF['Q5M10']
DF['Q6']=DF['Q6aM4']+DF['Q6bM6']
DF=DF.fillna(0)
DF
```

### âœ… Conclusion

After exploring the dataset, here are some key insights:

1. **Participation Variability**:
   - Not all students answered every question.
   - For example, `Q4bM7` has only 26 responses out of 86, while `Q1aM4` has 85 â€” suggesting either optional sections or low attempt rates on harder/optional questions.

2. **Performance Trends**:
   - Questions with higher attempts (`Q1aM4`, `Q1bM6`) were more likely core or compulsory.
   - Some students scored well in certain parts but left others blank, indicating selective answering.

3. **Total Score Correlation**:
   - There is likely a correlation between attempted questions and total marks â€” students attempting more questions or high-value ones (`Q5M10`, `Q6bM6`) likely scored higher.

4. **Missing Data**:
   - Substantial missing data in later questions (`Q6aM4`, `Q6bM6`) could either indicate time constraints, optionality, or student fatigue.
